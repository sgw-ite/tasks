```python
import numpy as np
import math
import matplotlib.pyplot as plt

class Pso():
    def __init__(self,dim,Xbound,Vbound,size,c1=2,c2=2,w=1,max_iter=10000):
        self.dim = dim
        self.Xbound = Xbound
        self.Vbound = Vbound
        self.size = size
        self.max_iter = max_iter
        self.c1 = c1
        self.c2 = c2
        self.w = w
        self.X = np.zeros((self.size,self.dim))
        self.V = np.zeros((self.size,self.dim))
        self.pbestX = np.zeros((self.size,self.dim))
        self.gbestX = np.zeros((1,self.dim))
        self.pbesty = np.zeros(self.size)
        self.gbesty = 1e5
        self.y = 1e5
    def Y(self,X):
        y = np.sum(X[i]**2-math.cos(2*math.pi*X[i]) for i in range(self.dim))
        return y
    def init_individual(self):
        for xx in range(self.size):
            for yy in range(self.dim) :
                self.X[xx][yy] = (np.random.random(1) - 0.5) * 5.12 * 2
                self.V[xx][yy] = (np.random.random(1) - 0.5) * 5.12 * 2
            self.pbestX[xx] = self.X[xx]
            self.y = self.Y(self.X[xx])
            self.pbesty[xx] = self.y
            if self.y < self.gbesty:
                self.gbesty = self.y
                self.gbestX = self.X[xx]
        return self
    def iterator(self):
        yhistory = []
        Vhistory = np.zeros((self.size,self.dim))
        Xhistory = np.zeros((self.size,self.dim))
        partition = np.zeros((1,self.dim))
        for n in range(self.max_iter):
            self.r1 = np.random.random(1)
            self.r2 = np.random.random(1)
            for i in range(self.size):
                self.y = self.Y(self.X[i])
                if self.y < self.pbesty[i]:
                    self.pbesty[i] = self.y
                    self.pbestX[i] = self.X[i]
                if self.y < self.gbesty:
                    self.gbesty = self.y
                    self.gbestX = self.X[i]
                self.V[i] = self.w * self.V[i] + self.c1 * self.r1 * (self.pbestX[i] - self.X[i]) + \
                            self.c2 * self.r2 * (self.gbestX - self.X[i])
            self.X[i] = self.X[i] + self.V[i]
            for j in range(self.dim):
                if abs(self.V[i][j]) >= self.Vbound :
                    self.V[i][j] = self.Vbound
            for j in range(self.dim):
                if abs(self.X[i][j]) >= self.Xbound :
                    self.X[i][j] = self.Xbound
            yhistory.append(self.gbesty)
            if n>9990:
                Vhistory = np.hstack([Vhistory,self.V])
            if n>9990:
                Xhistory = np.vstack([Xhistory,self.X,partition])
        return yhistory
```

```python
mypso = Pso(dim = 5,size = 5,Xbound = 5.12,Vbound = 5.12)
mypso.init_individual()
answer = mypso.iterator()
```

```python
yhistory = answer
N = [i for i in range(10000)]
plt.plot(N,yhistory)
plt.xlabel('iter')
plt.ylabel('besty')
plt.legend()
plt.show()
```
