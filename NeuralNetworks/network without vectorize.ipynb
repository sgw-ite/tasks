```python
class biasnode():
    
    '''
    偏置项的节点。储存连接权重以及value。
    -----
    params:
    N: 下一层节点个数(不包括偏置节点)
    ph: 每一个方向的权重的history
    -----
    '''
    
    def __init__(self, N=0):
        self.value = 1
        self.w = np.array([np.random.random() for i in range(N)])
        self.ph = []
        for i in range(N):
            self.ph.append([])
    def activate(self):
        #无实际意义，仅为了统一
        pass
```

```python
class node_sigmoid():
    
    '''
    结点类，用于储存相关信息
    -----
    params:
    input: 从上一个层中每一个节点传过来的数据的和
    value: input经过激活函数处理
    w[i]: forward中，对下一层编号为i的节点的权重
    N: 下一层节点个数(不包括偏置节点)
    error_term: 残差
    ph: 每一个方向的权重的history
    -----
    '''
    
    def __init__(self, N=0):
        self.input = None
        self.value = None
        self.w = np.array([np.random.random() for i in range(N)])
        self.error_term = None
        self.ph = []
        for i in range(N):
            self.ph.append([])
    
    def activate(self):
        self.value = 1 / (1 + np.exp(-self.input))
    
    def backward(self):
        return self.value * (1 - self.value)
```

```python
class layer_sigmoid():
    
    '''
    层类，用于储存节点，完成forward,backward工作
    -----
    params:
    N: 下一层节点个数(不包括偏置节点)
    node_num: 节点个数(不包括偏置节点)
    -----
    '''
    
    
    def __init__(self, N=0, node_num=2, bias='True'):
        self.node_num = node_num
        self.bias = bias
        if self.bias=='Ture':
            self.nodes = []
            for i in range(self.node_num):
                self.nodes.append(node_sigmoid(N))
            self.nodes.append(biasnode(N))
        else: 
            self.nodes = []
            for i in range(self.node_num):
                self.nodes.append(node_sigmoid(N))
    
    def forward(self):
        
        '''
        仅用于激活层内每一节点
        '''
           
        for i in self.nodes:
            i.activate()
```

```python
class FC():
    
    '''
    网络，进行整体的forward，backward工作
    -----
    params:
    -----
    '''
    
    
    def __init__(self, X_train, Y_train, layer_num=3, eta=2, lamda=1, outlayer_node_num=1):
        self.outlayer_node_num = outlayer_node_num
        self.eta = eta
        self.lamda = lamda
        self.layer_num = layer_num
        self.X_train = X_train.copy()
        self.Y_train = Y_train.copy()
        self.layers = []
        self.biter = 0
        for i in range(self.layer_num):
            if i==0:
                self.layers.append(layer_sigmoid(node_num=self.X_train.shape[1], N=2))
            elif ((i!=(self.layer_num-1))&(i!=(self.layer_num-2))):
                self.layers.append(layer_sigmoid(node_num=2, N=2))
            elif (i==(self.layer_num-2)):
                self.layers.append(layer_sigmoid(node_num=2, N=1))
            else:
                self.layers.append(layer_sigmoid(node_num=outlayer_node_num, N=0, bias='False'))
                
    
    def forward(self, x_train): #现在还没有向量化。。。鸡肋
        
        for i in range(self.X_train.shape[1]): #对输入层的每一个节点赋值value
            self.layers[0].nodes[i].value = x_train[i]
            
        for i in range(self.layer_num-1): #对除了输入层的每一层：
            for j in range(len(self.layers[i+1].nodes)): #每一个节dian
                aaa = sum([(self.layers[i].nodes[k].value * self.layers[i].nodes[k].w[j]) for k in range(len(self.layers[i].nodes))])                  
                self.layers[i+1].nodes[j].input = aaa #获得input
            self.layers[i+1].forward() #获得激活值
            
        #y_pred_i = np.array([self.layers[-1].nodes[i].value for i in range(self.outlayer_node_num)]) #获得x_train对应的y_pred
        #np.vstack((self.y_pred, y_pred_i.reshape(1,-1))) #将每一个y_pred收集起来
        
        return self
    
    
    def backward(self): #没有向量化
        
        for m in range(self.Y_train.shape[0]): #计算每一个样本的损失函数对权重的偏导数
            
            y_train = self.Y_train[m]
            x_train = self.X_train[m]
       
            #更新value、input
            self.forward(x_train)
            
            # 计算输出层残差
            for i in range(len(self.layers[-1].nodes)):
                self.layers[-1].nodes[i].error_term = ((self.layers[-1].nodes[i].value - y_train[i]) * 
                                                       self.layers[-1].nodes[i].backward())
            # 计算隐藏层残差
            for l in range(1,self.layer_num): #从第一层开始，除了输出层，每两层之间：
                for i in range(len(self.layers[self.layer_num-1-l].nodes)): #的每一个节点(包括偏置项)：
                    self.layers[self.layer_num-1-l].nodes[i].error_term = 0
                    for j in range(self.layers[self.layer_num-l].node_num): #对下一层每一个节点(不包括偏置项)：
                        self.layers[self.layer_num-1-l].nodes[i].error_term += (
                                                                 (self.layers[self.layer_num-1-l].nodes[i].w[j] * 
                                                                  self.layers[self.layer_num-l].nodes[j].error_term))
                    self.layers[self.layer_num-1-l].nodes[i].error_term *= (
                                                                 self.layers[self.layer_num-1-l].nodes[i].backward())
                        #计算残差
            
            # 计算对应的偏导数
            for l in range(1,self.layer_num): #从第一层开始，每两层之间：
                for i in range(len(self.layers[self.layer_num-1-l].nodes)): #的每一个节点(包括偏置项)：
                    for j in range(self.layers[self.layer_num-l].node_num): #对下一层每一个节点(不包括偏置项)：
                        self.layers[self.layer_num-1-l].nodes[i].ph[j].append(
                                                        self.layers[self.layer_num-l].nodes[j].error_term * 
                                                        self.layers[self.layer_num-1-l].nodes[i].value)
                    #在其对下一节点的偏导数历史中加入该偏导数
            
        #更新w
        for l in range(1,self.layer_num): #从第一层开始，每两层之间：
            for i in range(len(self.layers[self.layer_num-1-l].nodes)): #的每一个节点(包括偏置项)：
                for j in range(self.layers[self.layer_num-l].node_num): #对下一层每一个节点(不包括偏置项)：
                    self.layers[self.layer_num-1-l].nodes[i].w[j] = (
                                                    self.layers[self.layer_num-1-l].nodes[i].w[j] - (self.eta * 
                                                    (np.mean(self.layers[self.layer_num-1-l].nodes[i].ph[j]) + 
                                                    (self.lamda * self.layers[self.layer_num-1-l].nodes[i].w[j]))))
                    
                    self.biter += 1
                    if self.biter % 20 == 0:
                        print('1', self.eta * (np.mean(self.layers[self.layer_num-1-l].nodes[i].ph[j]) + 
                                          (self.lamda * self.layers[self.layer_num-1-l].nodes[i].w[j])))
                        
        for l in range(1,self.layer_num): #从第一层开始，每两层之间：
                for i in range(len(self.layers[self.layer_num-1-l].nodes)): #的每一个节点(包括偏置项)：
                    for j in range(self.layers[self.layer_num-l].node_num):
                        self.layers[self.layer_num-1-l].nodes[i].ph[j] = []
```

```python
class NeuralNetworkClassifier():
    
    def __init__(self):
        pass
    
    def _J(self, fc, Y_pred):
        w2 = 0
        for l in range(fc.layer_num):
            for i in range(len(fc.layers[l].nodes)):
                w2 += np.sum((fc.layers[l].nodes[i].w)**2)
        print('t', self.Y_train[:7,:])
        print('p', Y_pred[:7,:])
        J = np.sum((Y_pred - self.Y_train)**2) * 0.5 / (Y_pred.shape[0])  + 0.5 * fc.lamda * w2
        return J
        
    def fit(self, X_train, Y_train, epsilon=1e-4):
        self.X_train = X_train
        self.Y_train = Y_train
        fc = FC(self.X_train, self.Y_train)
        _iter = 0
        
        while(True):
            fc2 = fc
            Y_pred2 = np.ones((1,self.Y_train.shape[1]))
            for i in range(self.X_train.shape[0]):
                fc2.forward(self.X_train[i])
                y_pred_i = np.array([fc2.layers[-1].nodes[i].value for i in range(fc2.outlayer_node_num)])#获得x_train对应的y_pred
                #print(y_pred_i)
                Y_pred2 = np.vstack((Y_pred2, y_pred_i.reshape(1,-1))) #将每一个y_pred收集起来
            Y_pred2 = np.delete(Y_pred2,0,0)
            fc.backward()
            Y_pred = np.ones((1,self.Y_train.shape[1]))
            for i in range(self.X_train.shape[0]):
                fc.forward(self.X_train[i])
                y_pred_i = np.array([fc.layers[-1].nodes[i].value for i in range(fc.outlayer_node_num)]) #获得x_train对应的y_pred
                Y_pred = np.vstack((Y_pred, y_pred_i.reshape(1,-1))) #将每一个y_pred收集起来
            Y_pred = np.delete(Y_pred,0,0)
            #print('2',Y_pred.shape, Y_pred2.shape)
            delta = self._J(fc, Y_pred) - self._J(fc2, Y_pred2)
            if _iter%10 == 0:
                print(delta)
            if abs(delta) < epsilon:
                print(Y_pred[:7,:])
                break
            _iter += 1
       
        self.fc = fc
        
        return self
    
    def predict(self, X_train):
        
        Y_pred = np.ones((1,self.Y_train.shape[1]))
        for i in range(self.X_train.shape[0]):
            self.fc.forward(self.X_train[i])
            y_pred_i = np.array([self.fc.layers[-1].nodes[i].value for i in range(self.fc.outlayer_node_num)])#获得x_train对应的y_pred
            #print(y_pred_i)
            Y_pred = np.vstack((Y_pred, y_pred_i.reshape(1,-1))) #将每一个y_pred收集起来
        Y_pred = np.delete(Y_pred,0,0)
        
        return Y_pred
```
